# MatPlotLib.PyPlot

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

## Scatter Plot

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y, c="#223BDC", alpha=0.8, marker="X")
plt.xlabel("This is my X-axis")
plt.ylabel("This is my Y-axis")
plt.title("This is my X vs Y")
plt.grid()
```

```python
x = np.linspace(1,10 , 100)
y = np.sin(x)
```

```python
x
```

```python
y
```

## Line plot/graph

```python
# Line Graph
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y, "--r")
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my sin wave graph")
plt.grid()
```

## Bar plot/graph

```python
x = ["a", "b", "c", "d", "e"]
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y, color = '#7E10D5')
plt.xlabel("representing my categorical data")
plt.ylabel("representing my numerical data")
plt.title("this is my bar plot")
plt.grid()
```

### Bar plot on horizontal axis (i.e categories on y-axis and numbers on x-axis)

```python
plt.figure(figsize=(5,3))
plt.barh(x,y, color = '#25e4d1')
plt.xlabel("representing my numerical data")
plt.ylabel("representing my categorical data")
plt.title("this is my horizontal bar plot")
plt.grid()
```

## Histogram

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
plt.hist(data)
```

```python
plt.figure(figsize=(6,3))
plt.hist(data, color="#700EBE")
plt.xlabel("representing unique value")
plt.ylabel("representing frequency")
plt.title("Hist plot example")
plt.grid()

# to save this graph/plot
plt.savefig("test_hist.png")
```

## 3-D plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.scatter(x,y,z, c='#9327E8',alpha = .9)
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.show()
```
